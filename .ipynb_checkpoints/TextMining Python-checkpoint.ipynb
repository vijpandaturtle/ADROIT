```python
import numpy as np
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer 
from sklearn.feature_extraction import text
```

```python
res = pd.read_csv("dataset.csv", sep=";")
res.head()
```

```python
#Printing all column names 
# for column in res.columns.to_list():
#     print("FEATURES : ", column)
```

```python
index_names = res[res['description'] == None].index   
res.drop(index_names, inplace = True) 
```

```python
descriptionCorpus = res['description']
descriptionCorpus.head()
#Converting all descriptions to lower case
descriptionCorpus = descriptionCorpus.str.lower()
#Removing numbers from description 
descriptionCorpus = descriptionCorpus.str.replace('\d+', '')
#Removing punctuation
descriptionCorpus = descriptionCorpus.str.replace('[^\w\s]','')
#Remove white space 
descriptionCorpus = descriptionCorpus.str.strip()
```

```python
#Removing stop words
stop_words = text.ENGLISH_STOP_WORDS
vectorizer = CountVectorizer(stop_words=stop_words, min_df=200)
descriptionCorpus = vectorizer.fit_transform(descriptionCorpus)
```

```python
#Shape of the document matrix
#print(descriptionCorpus)
descriptionCorpus = list(descriptionCorpus)
```

```python
newDataset = res
```

```python
newDataset['Matrix'] = pd.Series(descriptionCorpus, index=res.index)
newDataset.head()
```

```python
#Dropping labels and description from the dataframe 
newDataset = newDataset.drop(columns=['LABEL','description','.//MD5'])
```
